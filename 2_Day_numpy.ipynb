**indexing and slicing**

```python
x=np.arange(16).reshape(4,4)
x
```

```python
# indexing work on rows intially
x[0]
```

```python
x[-1]
```

```python
# select 14 from last row
x[-1,2]
```

```python
x[-1][3]
```

```python
x[:,:] #[row ,column]
```

```python
x[: , :2] # start:stop
```

```python
x
```

```python
#13,14
x[-1:,1:3]
```

```python
#2,3
#10,11
x[::2,2:]
```

```python
x.diagonal()
```

```python
# select corner element of an array
x[::3,::3]
```

```python
# replace corner element by 0
x[::3,::3]=0
x
```
